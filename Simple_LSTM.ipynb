```python
import numpy as np
from tensorflow.python.keras.models import Sequential
from tensorflow.python.keras.layers import LSTM, Dense
```

```python
X = []
Y = []
for i in range(5, 20):
    X.append([[j] for j in range(i - 5, i)])
    Y.append([i])
```

```python
X = np.array(X)
Y = np.array(Y)
```

```python
print(X.shape)
print(Y.shape)
print(X)
print(Y)
```

```python
# LSTM 모델 구성
# LSTM 레이어 50개층, Dense 레이어 1개층으로 구성된 모델
model = Sequential()
model.add(LSTM(50, activation='relu', input_shape=(5, 1)))
model.add(Dense(1, activation='linear'))
model.compile(loss='mse', optimizer='adam', metrics=['accuracy'])
```

```python
# 모델 학습
hist = model.fit(X, Y, epochs=100, batch_size=1)
```

```python
import matplotlib.pyplot as plt
plt.title('Loss')
plt.plot(hist.history['loss'])
plt.xlabel('epoch')
plt.ylabel('loss')
plt.show()
```

```python
# 다음 값을 예측
next_value = np.array([[[16], [17], [18], [19], [20]]])
# next_value = np.array([[[1], [2], [3], [4], [5]]])
predicted_value = model.predict(next_value)
# print(f"predict value: {predicted_value[0][0]}")
print(f"predict value: {predicted_value[0][0]}")
```
