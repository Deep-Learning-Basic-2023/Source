```python
dataframe = pandas.read_csv('/content/drive/MyDrive/Colab Notebooks/DL_Lec_2023/xor_data.txt', header=None)
dataset = dataframe.values
X_data = dataset[:, 0:2]
Y_data = dataset[:,2]
```

```python
print(X_data)
print(Y_data)
```

```python
model = Sequential()
model.add(Dense(2, input_dim=2, activation='relu', name='dense01'))
model.add(Dense(8, activation='relu', name='dense02'))
model.add(Dense(1, activation='sigmoid', name='dense03'))
```

```python
model.summary()
```

```python
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
model.fit(X_data, Y_data, epochs=2000)
```

```python
result = model.evaluate(X_data, Y_data)
```

```python
print(int(model.predict(np.array([[0, 0]])) > 0.5))
print(int(model.predict(np.array([[1, 0]])) > 0.5))
print(int(model.predict(np.array([[0, 1]])) > 0.5))
print(int(model.predict(np.array([[1, 1]])) > 0.5))
```
